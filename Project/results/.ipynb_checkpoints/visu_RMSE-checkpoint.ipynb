```python
from numpy import loadtxt
data_lidar = loadtxt('02-07-2019 22_23_50_RMSE_LIDAR.txt', delimiter="\n", unpack=False)
data_radar = loadtxt('02-07-2019 22_25_44_RMSE_RADAR.txt', delimiter="\n", unpack=False)
data_fusion = loadtxt('02-07-2019 22_11_39_RMSE_FUSION.txt', delimiter="\n", unpack=False)
```

```python
data_fusion_px = data_fusion[::4]
data_fusion_py = data_fusion[1::4]
data_fusion_vx = data_fusion[2::4]
data_fusion_vy = data_fusion[3::4]
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = [30, 15]
fig, ax = plt.subplots()
ax.plot(data_fusion_px)
ax.plot(data_fusion_py, 'r')

ax.legend(['PX', 'PY'], prop={'size': 30})
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = [30, 15]
fig, ax = plt.subplots()
ax.plot(data_fusion_vx, 'g')
ax.plot(data_fusion_vy, 'k')

ax.legend(['VX', 'VY' ], prop={'size': 30})
```

```python
N = 1600
data_lidar = data_lidar[:N]
data_radar = data_radar[:N]
data_fusion = data_fusion[:N]
```
